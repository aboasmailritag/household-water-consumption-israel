How has household water consumption changed over time in Israel? (1998-2022)

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Year": [1998, 2000, 2005, 2010, 2015, 2020, 2022],
    "Household_NonAgri_Consumption": [672, 662, 715, 689, 893, 1030, 1091]
}

df = pd.DataFrame(data)
df

```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Year"], df["Household_NonAgri_Consumption"], marker="o")
plt.xlabel("Year")
plt.ylabel("Water Consumption (million m³)")
plt.title("Household Water Consumption in Israel (1998–2022)")
plt.grid(True)
plt.show()
```

Household water consumption in Israel shows an overall increasing trend between 1998 and 2022 with a significant rise after 2015 indicating structural changes in water usage patterns.

```python
before_2015 = df[df["Year"] < 2015]["Household_NonAgri_Consumption"].mean()
after_2015  = df[df["Year"] >= 2015]["Household_NonAgri_Consumption"].mean()

before_2015, after_2015
```

Average household water consumption increased from approximately 665 million m³ before 2015 to about 793 million m³ after 2015 indicating a structural change in consumption patterns.
